```python
from urllib.request import urlopen
from bs4 import BeautifulSoup

url = 'https://adotar.com.br/ms-campo-grande/cao/adocao/biscoito-e-nemo-macho-medio/208783'

response = urlopen(url)
html = response.read()

soup = BeautifulSoup(html, 'html.parser')

soup
```

```python
contact = soup.find('a',{"id":"mailprop"})
email = contact['href']
email
phone = contact.findNextSibling().find('a').getText()
print(email)
print(phone)
```
